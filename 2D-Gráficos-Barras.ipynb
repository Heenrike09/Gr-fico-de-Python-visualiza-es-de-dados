```python
# importando as bibliotecas e classes do projeto
import numpy as np # trabalha com objetos vetor e matriz multidimencional
import matplotlib.pyplot as plt # gerar os graficos de visualização
```

```python
# criando os campos Descritivos (Rótulo labels)
labels = ["A1","A2","A3","A4","A5"]
```

```python
# criando os valores dos campos (Métrica do Gráfico)
valores = [10,34,56,78,12] 
```

```python
plt.bar(labels,valores)
```

```python
plt.barh(labels,valores)
```

```python
# atrinuidos novos aos eixos do gráfico (x,y)
plt.bar(labels,valores)
plt.xlabel('Valor do produto')
plt.ylabel('Valor do produto')
```

```python
# alterar as cores do elemento do gráfico
plt.bar(labels,valores,color='red')
plt.xlabel('Valor do produto')
plt.ylabel('Valor do produto')
```

```python
# alterar as cores do elemento do gráfico
plt.barh(labels,valores,color='red')
plt.xlabel('Valor do produto')
plt.ylabel('Valor do produto')
```

```python
# colando os valores existentes em casa tipo de como rótulo de dados
plt.bar(labels,valores,color='red')
plt.xlabel('Valor do produto')
plt.ylabel('Valor do produto')

for x,y in zip(labels,valores):
    plt.text(x, y+0.05, '%d' % y, ha='center' , va= 'bottom')
```

```python
# Criando um novo vetor chamado valores2
valores2 = [8,30,60,21,30]
```

```python
# Plotar o gráfico  com os dois gráficos e elementos de valores juntos
largura = 0.4
pos1 = np.arange(len(valores))
pos2 = [x + largura for x in pos1]

plt.bar(pos1,valores, color='red', width=largura, label='Barra 1')
plt.bar(pos2, valores2, color='blue', width=largura, label='Barra 2')
plt.legend()
```

```python
plt.bar(pos1,valores,label = "Barra 1",color="green")
plt.bar(pos1,valores2,label = "Barra 2", bottom = valores)
```

```python
plt.barh(pos1,valores,label = "Barra 1")
plt.barh(pos1,valores2,label = "Barra 2",left = valores)
```
