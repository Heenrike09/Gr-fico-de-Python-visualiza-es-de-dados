```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,7,14)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Números de prodtos')
plt.ylabel('Valores')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.plot(X,Z,label='$Z(X)=e^{(x/2)}$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend(loc='lower right')
plt.show()
```

```python
X = np.linspace(0,2.0*np.pi,361)
```

```python
Y = np.cos(X/2)
Z = np.sin(X/4)
```

```python
plt.plot(X,Y,label='$\cos(X/2)$',color='red')
plt.plot(X,Z,label='$\sin(X/4)$',color='green',linestyle='--')
plt.legend()
```
