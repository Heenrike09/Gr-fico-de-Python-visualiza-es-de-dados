```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,9,17)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.subplot(2,1,1)
plt.plot(X,Y)
plt.subplot(2,1,2)
plt.plot(X,Z)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(X,Y)
plt.xlabel('Gráu de crecimento')
plt.ylabel('Dias')
plt.subplot(2,1,2)
plt.plot(X,Z)
plt.xlabel('Grau de crecimento')
plt.ylabel('Dias')
plt.tight_layout()
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(X,Y)
plt.xlabel('Grau de crecimento')
plt.ylabel('Dias')
plt.subplot(1,2,2)
plt.plot(X,Z)
plt.xlabel('Gráu de crecimento')
plt.ylabel('Dias')
plt.tight_layout()
plt.show()
```

```python
fig,ax = plt.subplots(nrows=1,ncols=2)
ax[0].plot(X,Y)
ax[0].set_xlabel('Grau de crecimento')
ax[0].set_xlabel('Dias')
ax[1].plot(X,Z)
ax[1].set_xlabel('Gra de crecimento')
ax[1].set_xlabel('Dias')
plt.tight_layout()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1)
ax[0].plot(X,Y)
ax[0].set_xlabel('Grau de crecimento')
ax[0].set_xlabel('Dias')
ax[1].plot(X,Z)
ax[1].set_xlabel('Gra de crecimento')
ax[1].set_xlabel('Dias')
plt.tight_layout()
```

```python
Y = np.power(X-2,2.0)
Z = np.exp(X/2.0)
W = np.power(X,3)
L = np.power(X,0.5)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2)
ax[0][0].plot(X,Y)
ax[0][0].set_xlabel('Eixo X')
ax[0][0].set_ylabel('Eixo Y')
ax[0][1].plot(X,Z)
ax[0][1].set_xlabel('Eixo X')
ax[0][1].set_ylabel('Eixo Z')

ax[1][0].plot(X,W)
ax[1][0].set_xlabel('Eixo X')
ax[1][0].set_ylabel('Eixo W')
ax[1][1].plot(X,L)
ax[1][1].set_xlabel('Eixo X')
ax[1][1].set_ylabel('Eixo L')
plt.tight_layout()
```
