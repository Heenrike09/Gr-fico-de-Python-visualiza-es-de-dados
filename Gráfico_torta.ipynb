```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
labels = ["A1","A2","A3","A4","A5","A6"]
```

```python
valores = [10,48,39,52,23,12]
```

```python
plt.pie(valores,labels=labels)
plt.show()
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%')
plt.show()
```

```python
explode = [0,0.1,0,0,0,0.1]
```

```python
plt.pie(valores,labels=labels, autopct='%1.1f%%',explode=explode)
plt.show()
```
