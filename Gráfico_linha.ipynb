```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,7,14)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.plot(X,Y)
plt.show()
```

```python
plt.plot(X,Y)
plt.xlabel('Número de prodtos')
plt.ylabel('Valores de produtos')
plt.show()
```

```python
plt.plot(X,Y)
plt.plot(X,Z)
plt.xlabel('Número de produtos')
plt.ylabel('Valores dos prodtos')
plt.title('Produto 1, Produto 2')
plt.show()
```

```python
plt.plot(X,Y,color='pink')
plt.plot(X,Z,color='green')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.show()
```

```python
plt.plot(X,Y,color='pink',label='valor 1')
plt.plot(X,Z,color='red',label='valor 2')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X), Z(X)')
plt.legend()
plt.show()
```
