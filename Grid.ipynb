```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(0,9,18)
```

```python
Y = np.power(X-2,2.0)
```

```python
Z = np.exp(X/2.0)
```

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Número de produtos')
plt.ylabel('Prospecto de vendas')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```

```python
plt.plot(X,Y,label='$Y(X)=(X-2)^2$')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.title('Y(X)')
plt.grid(linestyle = '--' , which = 'major' , color = 'grey')
plt.legend()
plt.show()
```
