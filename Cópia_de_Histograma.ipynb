```python
import numpy as npimport 
import matplotlib.pyplot as plt
import numpy as np

```

```python
X = np.random.rand(1500)
```

```python
plt.hist(X)
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.hist(X,bins=20)
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.title('Bins=20')
plt.show()
```

```python
plt.hist(X,bins=30)
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.title('Bins=30')
plt.show()
```

```python
plt.hist(X,bins=50,color='purple')
plt.xlabel('Valores')
plt.ylabel('Contagem')
plt.title('Bins=50')
plt.show()
```

```python
plt.hist(X,bins=40,color='purple', edgecolor='black')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Bins=40')
plt.show()
```

```python
X1 = np.random.normal(0,1,size = 800)
X2 = np.random.normal(3,1,size = 800)
X3 = np.random.normal(5,1,size = 800)
```

```python
plt.hist(X1,bins=10,color='purple')
plt.hist(X2,bins=10,color='blue')
plt.hist(X3,bins=10,color='green')
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Multiplos histogramas')
plt.show()
```

```python
X = np.random.rand(2000)
```

```python
plt.hist(X,bins=20,color='red', edgecolor='blue',cumulative=True)
plt.xlabel('X')
plt.ylabel('Contagem')
plt.title('Histograma cumulativo')
plt.show()
```
